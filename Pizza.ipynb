```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

data= {
    'size':['small', 'medium', 'large', 'large', 'medium'],
    'crust':['Thin', 'Pan', 'cheese burst', 'Pan', 'Thin'],
    'Is_Nonveg':[0, 0, 1, 1, 0],
    'Add Drink':[1, 0, 1, 1, 0],
    'Extra cheese':[0, 1, 1, 1, 0],
    'Label':['Basic veg', 'Deluxe veg', 'Chicken combo', 'Meat feast', 'Healthy veg'],

}
df=pd.DataFrame(data)
print(df)

from sklearn.preprocessing import LabelEncoder

le_size=LabelEncoder()
le_crust=LabelEncoder()

df['size_enc']= le_size.fit_transform(df['size'])
df['crust_enc']=le_crust.fit_transform(df['crust'])

x = df[['size_enc', 'crust_enc', 'Is_Nonveg', 'Add Drink', 'Extra cheese']]
y = df['Label']

model = DecisionTreeClassifier()
model.fit(x,y)

plt.figure (figsize=(12, 6))
plot_tree(model, feature_names=x.columns, class_names=model.classes_, filled=True)
plt.title("Pizza Decision Tree")
plt.show()
```
